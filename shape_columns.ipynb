```python
import pandas as pd
df = pd.read_json('sample_Data.json')
```

```python
print("Shape: ",df.shape)
print("Column Names: ", df.columns)
```

```python
#selecting any single column
print(df['name'])

#selecting multiple columns
subset = ['description','price']
print(df[subset])
```

```python
high_sales = df[df['price'] > 800]
print(high_sales.head(5))
```

```python
#Multiple Filters
filtered = df[(df['price'] > 900) & (df['category']=='Electronics')]
print(filtered)
```

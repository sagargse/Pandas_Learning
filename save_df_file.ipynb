```python
import pandas as pd
data={
    "name": ["sagar","Raj","Praful"],
    "age":[25,21,34],
    "city": ["Patna",'Mumbai','Nagpur']
}
```

```python
print(data)
df = pd.DataFrame(data)
print(df)
```

```python
# To save datframe into files

df.to_csv("output_csv.csv",index=False) #index false to remove the indexing
df.to_excel("output_excel.xlsx",index=False) 
df.to_json("output_json.json",index=False)
```

```python
# to read files
df = pd.read_csv("output_csv.csv")
print(df,"\n")
df = pd.read_excel("output_excel.xlsx")
print(df,"\n")
df = pd.read_json("output_json.json")
print(df,"\n")
```
